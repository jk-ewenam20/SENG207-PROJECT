## Filtering, Sorting and filling null values

```python
mean_fa = wine_data['fixed acidity'].mean()
mean_va = wine_data['volatile acidity'].mean()
mean_cl = wine_data['chlorides'].mean()
mean_fsd = wine_data['free sulfur dioxide'].mean()
mean_density = wine_data['density'].mean()
mean_quality = wine_data['quality'].mean()
```

```python
print(f'mean of fixed acidity: {mean_fa}')
print(f'mean of volatile acidity: {mean_va}')
print(f'mean of chlorides: {mean_cl}')
print(f'mean of free sulfur dioxide: {mean_fsd}')
print(f'mean of mean density: {mean_density}')
print(f'mean of quality: {mean_quality}')
```

```python
wine_data['fixed acidity'] = wine_data['fixed acidity'].fillna(mean_fa)
wine_data['volatile acidity'] = wine_data['volatile acidity'].fillna(mean_va)
wine_data['chlorides'] = wine_data['chlorides'].fillna(mean_cl)
wine_data['free sulfur dioxide'] = wine_data['free sulfur dioxide'].fillna(mean_fsd)
wine_data['density'] = wine_data['density'].fillna(mean_density)
wine_data['quality'] = wine_data['quality'].fillna(mean_quality)
```
